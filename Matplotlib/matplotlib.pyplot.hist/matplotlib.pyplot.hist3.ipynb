```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from math import *
from scipy.stats import norm

nor = np.random.normal(0,1,500) #ガウシアンに従う乱数500個発生
x = np.arange(-5,5,0.01)

plt.xlabel("x")
plt.ylabel("N")
plt.hist(nor,bins=20,range=(-4,4))
plt.plot(x,norm.pdf(x)*200,lw=3,color="r")
plt.show()
```
