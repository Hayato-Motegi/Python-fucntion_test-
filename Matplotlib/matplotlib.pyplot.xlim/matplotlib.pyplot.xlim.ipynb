# Matplotlib で軸の制限を設定する方法
## https://www.delftstack.com/ja/howto/matplotlib/how-to-set-limit-for-axes-in-matplotlib/#matplotlib-%E3%81%A7%E8%BB%B8%E3%81%AE%E7%AF%84%E5%9B%B2%E3%82%92%E8%A8%AD%E5%AE%9A%E3%81%99%E3%82%8B%E3%81%9F%E3%82%81%E3%81%AE-xlim-%E3%81%8A%E3%82%88%E3%81%B3-ylim

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10,500)
y=np.sin(2 * np.pi * x)+1

fig = plt.figure(figsize=(8, 6))
plt.plot(x,y)
plt.title("Setting range of Axes",fontsize=25)
plt.xlabel("x",fontsize=18)
plt.ylabel("1+sinx",fontsize=18)
plt.xlim(4,8)
plt.ylim(-0.5,2.5)
plt.show()
```
