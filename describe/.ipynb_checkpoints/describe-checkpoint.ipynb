# PandasでDataFrameの統計情報を表示するdescribe関数の使い方
## https://deepage.net/features/pandas-describe.html

```python
import pandas as pd
import numpy as np # 乱数生成のためにNumPyをインポートしておく。
```

```python
data = np.random.randint(-10,10,size=100).reshape(20,5)
```

```python
df = pd.DataFrame(data,columns=['A','B','C','D','E'])
```

```python
df # 数値の羅列だけだとよく分からない部分が多い
```

```python
df.describe() # describe関数でとりあえず概要表示
```

### 上からcount,mean,std,min,25%,50%,75%,maxと並んでいますが統計をやったことのある方にはすぐにピンと来る名称だと思います。 
### 日本語に直しますと順にデータ個数、平均、標準偏差、最小値、第一四分位数、第二四分位数、第三四分位数、最大値と並んでいます。
