```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # 3次元グラフをかくときに使います。

import math
```

```python

range_xy: list = range(-5 , 5, 1)
a: list = np.array(range_xy)
b: list = np.array(range_xy)
```

```python
print(f"a = {a}")
print(f"b = {b}")
```

```python
aa, bb = np.meshgrid(a,b)
```

```python
print(f"aa = {aa}")
print(f"bb = {bb}")
```

```python
data: list = aa**2+bb**2
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(aa, bb, data)
plt.show()
```
